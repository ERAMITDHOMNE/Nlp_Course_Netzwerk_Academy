```
import pdfx  #Web scraping mean to extract data from website ,any website , upsc WEBSITE DECLARED ,DATASET : TO GET DATA WHO CLEARED EXAM 
# extract data from govt website who declared UPSC result with web scraping 
pdf = pdfx.PDFx("http://europepmc.org/articles/PMC3001474?pdf=render") # wesite to pdf # go to url and then convert it into PDF

pdf
```

```
text = pdf.get_text() # assign extracted text data to variable text, We get whole data at once from any website 
# instead of click on particular link and get data available in pdf we are whole dat at once
print(text)
```

```

import spacy  # with the use of this library you can extract/ mining text data from text file,pdf file ,image file
# spacy added to my code for english text data.Then nlp variable we be asssign with Spacy.load module because we are working on english data
nlp = spacy.load("en_core_web_sm") # module we need to add to program becaue we are working on english text data
doc = nlp(text)# calling spacy library module assign to nlp variable and text varible holding text data of my website
```

```
doc # we are working on spacy library.The motive behind spacy to aware about different text mining library according to our easyness and depand dataset,
# for Image data spacy library works well #here we did same thing as above but with the use of spacy library instead of pdfx library tool 
```

```
import pandas as pd

cols = ("text", "lemma", "POS", "explain", "stopword") # I want output to be look like this colunm row format
rows = [] # empty list

for t in doc: # docs variable holding text data extracted from pdf file, for every word (t) present in doc varible or in webste,pdf it will loop for every variable
    row = [t.text, t.lemma_, t.pos_, spacy.explain(t.pos_), t.is_stop] # t <=1000 , I am working in oracle R & D # work # I am eating and I am watching tv
    rows.append(row) # How was the day??

df = pd.DataFrame(rows, columns=cols)
df
```

```
for ent in doc.ents: # 9912 words 0- > words  ent of words
    print(ent.text, ent.label_) # ent.label = Name entity, number, person name,date # advantage of spacy entity descrition,entity mining with ent function
```
